# Analyzing The League Data

```python
#Importing Libraries
import numpy as np #To handle Mathematical calculations
import matplotlib.pyplot as plt #To plot charts 
%matplotlib inline 
import pandas as pd #TO import and manage datasets
import glob   #TO find pathnames
import os #OS functions that are used to perform OS-based tasks
```

```python
#path = r'D:\DreamElevenScouting\Datasets\Whoscored\Points-Table'
path = r'../Datasets/Whoscored/Points-Table'
folder = glob.glob(path + "/*.csv")

list = []

for file in folder:
#     df = pd.read_csv(folder\filename, index_col=None, header=0)
    individualTeamsDataframe = pd.read_csv(file)
    league = os.path.basename(file).split('-')[0]
    season = os.path.basename(file).split('-')[3]
    individualTeamsDataframe['League'] = league
    individualTeamsDataframe['Season'] = int(season)
    list.append(individualTeamsDataframe)

# combinedDataframe = pd.concat(list, axis=0, ignore_index=True)
combinedTeamsDataframe = pd.concat(list)

# dataset = pd.read_csv('D:\DreamElevenScouting\Datasets\Whoscored\Points-Table\LaLiga-Points-Table-2018-2019.csv')
# league = 'LaLiga'
```
